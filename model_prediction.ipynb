```python
import numpy
import os
import pandas as pd
from tensorflow.keras.preprocessing import image
from PIL import ImageTk, Image
from tensorflow.keras.models import load_model
```

```python
model = load_model(filepath="model1_catsVSdogs_10epoch.h5")
```

```python
img_path = "./dogs-vs-cats/test/1.jpg"
```

```python
def predict_dog_or_cat(path, model):
    image = Image.open(path)
    image = image.resize((128,128))
    image = numpy.expand_dims(image, axis=0)
    image = numpy.array(image)
    image = image/255
    pred = model.predict_classes(image)[0]
    return pred
```

```python
predict_dog_or_cat(img_path, model)
```

```python
test_files = os.listdir("dogs-vs-cats/test/")
```

```python
test_labels = []
for file in test_files:
    test_labels.append(predict_dog_or_cat(os.path.join("./dogs-vs-cats/test/",file), model))
```

```python
test_labels
```

```python
file = []
for img in test_files:
    file.append(img.split(".")[0])
```

```python
df_test = pd.DataFrame({
    "id": file,
    "label": test_labels
})
```

```python
df_test.sort_values(["id"],inplace=True)
```

```python
df_test.to_csv("submission1.csv")
```
